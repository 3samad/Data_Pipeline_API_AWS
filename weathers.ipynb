```python
def get_weather(cities):

    API_key = "DO_NOT_FORGET_TO_PUT_YOUR_KEY_HERE"

    now = datetime.now()

    weather_dict = {"city": [],
                    "country": [],
                    "forecast_time": [],
                    "weather_outlook": [],
                    "weather_detailed": [],
                    "temperature": [],
                    "feels_like_temperature": [],
                    "humidity": [],
                    "wind_speed": [],
                    "risk_of_rain": [],
                    "amount_of_rain": [],
                    "amount_of_snow": [],
                    "info_retrieved_at": []}

    for city in cities:
        url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric"
        response = requests.get(url)
        wea_resp = response.json()

        for w in wea_resp["list"]:
            weather_dict["city"].append(wea_resp["city"]["name"])
            weather_dict["country"].append(wea_resp["city"]["country"])
            weather_dict["forecast_time"].append(w["dt_txt"])
            weather_dict["weather_outlook"].append(w["weather"][0]["main"])
            weather_dict["weather_detailed"].append(w["weather"][0]["description"])
            weather_dict["temperature"].append(w["main"]["temp"])
            weather_dict["feels_like_temperature"].append(w["main"]["feels_like"])
            weather_dict["humidity"].append(w["main"]["humidity"])
            weather_dict["wind_speed"].append(w["wind"]["speed"])
            weather_dict["risk_of_rain"].append(w["pop"])

            try:
              weather_dict["amount_of_rain"].append(w["rain"]["3h"])
            except:
              weather_dict["amount_of_rain"].append("0")
            try:
              weather_dict["amount_of_snow"].append(w["snow"]["3h"])
            except:
              weather_dict["amount_of_snow"].append("0")

            weather_dict["info_retrieved_at"].append(now.strftime("%d/%m/%Y %H:%M:%S"))

    return pd.DataFrame(weather_dict)
```

```python
# Calling the function by taking a list of cities as input

cities = ["Leipzig", "Dresden", "Frankfurt"]

weathers = get_weather(cities)
weathers
```
